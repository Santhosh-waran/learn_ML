```python
from sklearn.datasets import load_digits
from sklearn import metrics
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sas 
%matplotlib inline 
df = load_digits()
```

```python
print(df.data.shape)
print(df.target.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming df.data is the image data and df.target contains the labels
plt.figure(figsize=(20, 4))
for index, (img, label) in enumerate(zip(df.data[:5], df.target[:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(img, (8, 8)), cmap=plt.cm.gray)
    plt.title(f"Label: {label}")

plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.data,df.target,test_size=0.23,random_state = 2)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(df.data)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression(max_iter=1000)
# np.reshape(-1,-1)
logr = logr.fit(x_train,y_train)
```

```python
y_pred = logr.predict(x_test)
y_pred.shape
```

```python
score = logr.score(x_test,y_test)
print(score)
```

```python
cm = metrics.confusion_matrix(y_test,y_pred)
plt.figure(figsize=(9,9))
sns.heatmap(cm ,annot=True,cmap='Blues_r', fmt ='.3f',linewidth=5,square= True)
plt.xlabel("x lable")
plt.ylabel("y lable")  
all_sample_title = 'accuracy score :{0}'.format(score)
plt.title(all_sample_title,size=15)
# print(cm)
```
